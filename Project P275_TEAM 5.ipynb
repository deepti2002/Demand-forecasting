## Use the Augmented Dickey-Fuller (ADF) Test: 

```python
#The ADF test is a statistical test used to determine whether a time series is stationary or not. 
```

```python
from statsmodels.tsa.stattools import adfuller

# result = adfuller(data['sales'])

# print('ADF Statistic:', result[0])
# print('p-value:', result[1])
# print('Critical Values:', result[4])
```

##### As we have a large dataset, Using a sample or aggregating the data can be helpful to speed up the ADF test



### Aggregating the Data:
###### Aggregating involves summarizing the data over larger time intervals. For example, if data is recorded on a daily basis, we can aggregate it to weekly or monthly intervals. This reduces the number of data points while retaining the overall trends and patterns.

```python
data
```

```python
aggregated_data = data.resample('W-Mon').sum()  # Aggregating to weekly intervals

result = adfuller(aggregated_data['sales'].dropna())

print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
```

###### Based on the ADF test results, the p-value (0.2847875125403818) is greater than the significance levels of 1%, 5%, and 10%. This indicates that you fail to reject the null hypothesis. The null hypothesis in the ADF test is that the data contains a unit root (it is non-stationary). The higher p-value suggests that the data might indeed be non-stationary.

# Make Data Stationary:

## Differencing:

```python
# Assuming your time series data is in a DataFrame called 'data'
aggregated_data['diff_sales'] = aggregated_data['sales'] - aggregated_data['sales'].shift(1)
aggregated_data['diff_sales'].dropna(inplace=True)
```

```python
# Assuming your differenced, aggregated data is in a DataFrame called 'aggregated_data'
result_diff_aggregated = adfuller(aggregated_data['diff_sales'].dropna())

print('ADF Statistic:', result_diff_aggregated[0])
print('p-value:', result_diff_aggregated[1])
print('Critical Values:', result_diff_aggregated[4])
```

##### Based on the ADF test results, the p-value ( 0.000496054070068327) is below the significance levels of 1%, 5%, and 10%. This suggests that you can reject the null hypothesis, indicating that the differenced, aggregated data is stationary.

## But,this doesn't automatically imply that the entire original dataset is also stationary.
## it's important to consider the characteristics of the original data before making a final decision.

## We check with : 
## Rolling Statistics:
#### Calculate rolling statistics (e.g., rolling mean and rolling standard deviation) and plot them against the original data. If these statistics remain relatively constant over time, it suggests stationarity.

```python
# Assuming your time series data is in a DataFrame called 'data'
window = 30  # Choose the size of the rolling window

# Calculate rolling mean and rolling std
data['rolling_mean'] = data['sales'].rolling(window=window).mean()
data['rolling_std'] = data['sales'].rolling(window=window).std()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['sales'], label='Original Data')
plt.plot(data.index, data['rolling_mean'], label=f'Rolling Mean (Window {window})', color='red')
plt.plot(data.index, data['rolling_std'], label=f'Rolling Std (Window {window})', color='green')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Original Data and Rolling Statistics')
plt.legend()
plt.show()
```

```python
# Calculate rolling mean and rolling std for the original data
window = 30  # Choose the size of the rolling window
data['rolling_mean_original'] = data['sales'].rolling(window=window).mean()
data['rolling_std_original'] = data['sales'].rolling(window=window).std()

# Calculate the range of deviations in rolling mean and rolling std for the original data
rolling_mean_deviation_range_original = data['rolling_mean_original'].max() - data['rolling_mean_original'].min()
rolling_std_deviation_range_original = data['rolling_std_original'].max() - data['rolling_std_original'].min()
original_data_range = data['sales'].max() - data['sales'].min()

# Compare the magnitudes to the range of the original data
print(f"Rolling Mean Deviation Range (Original Data): {rolling_mean_deviation_range_original}")
print(f"Rolling Std Deviation Range (Original Data): {rolling_std_deviation_range_original}")
print(f"Original Data Range: {original_data_range}")

# Compare the magnitudes to assess the impact of deviations
if rolling_mean_deviation_range_original < original_data_range and rolling_std_deviation_range_original < original_data_range:
    print("Deviations for the original data are relatively small compared to the data range.")
else:
    print("Deviations for the original data might be significant compared to the data range.")
```

#### If the rolling mean and rolling standard deviation remain relatively constant or show minimal variation as they move through time, it's an indication that the data might be stationary. But, we are observing significant fluctuations, it suggests non-stationarity.

### To make a non-stationary time series data stationary, using Differencing

```python
# Assuming your time series data is in a DataFrame called 'data'
data['diff_sales'] = data['sales'] - data['sales'].shift(1)
data['diff_sales'].dropna(inplace=True)
```

```python
# Calculate rolling mean and rolling standard deviation
window_size = 7  # You can adjust the window size as needed
rolling_mean = data['diff_sales'].rolling(window=window_size).mean()
rolling_std = data['diff_sales'].rolling(window=window_size).std()

# Plot the differenced data, rolling mean, and rolling standard deviation
plt.figure(figsize=(12, 6))
plt.plot(data.index,data['diff_sales'], label='Differenced Sales', color='b')
plt.plot(data.index, rolling_mean, label='Rolling Mean', color='r')
plt.plot(data.index, rolling_std, label='Rolling Std Dev', color='g')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Differenced Data and Rolling Statistics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the range of deviations in rolling mean and rolling std
mean_deviation_range = rolling_mean.max() - rolling_mean.min()
std_deviation_range = rolling_std.max() - rolling_std.min()

# Calculate the range of your original data
original_data_range = data['diff_sales'].max() - data['diff_sales'].min()

# Compare the magnitude of deviations to the range of the original data
print("Range of Deviations in Rolling Mean:", mean_deviation_range)
print("Range of Deviations in Rolling Std:", std_deviation_range)
print("Range of Original Data:", original_data_range)
```

#### The calculations indicate that the deviations in the rolling mean and rolling standard deviation are relatively small compared to the range of our original data. This suggests that the achieved stationarity through the first differencing process is likely sufficient for our modeling and forecasting efforts.

```python
# Resample the data on a weekly basis, taking the mean value for each week
weekly_data = data['diff_sales'].resample('W').mean().dropna()
```

```python
# Perform the ADF test on the weekly data
result = adfuller(weekly_data)

# Extract and print test results
adf_statistic = result[0]
p_value = result[1]

print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {p_value}')

if p_value <= 0.05:
   print('Reject the null hypothesis. Weekly sampled data is stationary.')
else:
   print('Fail to reject the null hypothesis. Weekly sampled data is non-stationary.')
```

##### The goal of achieving stationarity is to make the data's statistical properties relatively stable over time, which helps in building accurate time series models.

##### The trend component represents the long-term movement or overall direction of the data, without the short-term fluctuations.

##### The residual component, also known as the remainder or error component, represents the variability in the data that remains after removing the trend and seasonal effects. It includes the random noise or irregular fluctuations that couldn't be explained by the trend and seasonality.

### 2. SARIMA MODEL

# Define SARIMA orders
#### p = 1  # Autoregressive order
#### d = 1  # Regular differencing order (already applied)
#### q = 2  # Moving Average order

### Triple Exponential Smoothing (Holt-Winters with Seasonality)
### 4.3 Parameter Interpretation:

#### It has three smoothing parameters: alpha (Î±), beta (Î²), and gamma (Î³). These parameters control the model's behavior and how it responds to different components of a time series, including level, trend, and seasonality.
#### Alpha (Î±) is primarily concerned with capturing the level or baseline of the time series.
#### Beta (Î²) focuses on modeling and capturing the trend in the data.
#### Gamma (Î³) deals with seasonality, capturing repeating patterns in the data.


```python
# Extract the smoothing parameters
alpha = fit.params['smoothing_level']  # Alpha (level smoothing)
beta = fit.params['smoothing_trend']   # Beta (trend smoothing)
gamma = fit.params['smoothing_seasonal']  # Gamma (seasonality smoothing)
```

```python
# Interpretation of the parameters
print(f"Alpha (Level Smoothing): {alpha:.4f}")
print(f"Beta (Trend Smoothing): {beta}")
print(f"Gamma (Seasonality Smoothing): {gamma:.4f}")
```

#### we can see, it is not giving proper results
#### we need to iteratively adjust your modeling approach, including the choice of initial parameter values and model specifications, until we achieve a satisfactory fit to our time series data.

```python
# Choose initial values for alpha, beta, and gamma
initial_alpha_values = [0.1, 0.3, 0.5]
initial_beta_values = [0.1, 0.2]
initial_gamma_values = [0.1, 0.2]
```

##### GRID SEARCH to systematically explore combinations of alpha,beta and gamma :

```python
import itertools

# Create a grid of parameter combinations
parameter_grid = list(itertools.product(initial_alpha_values, initial_beta_values, initial_gamma_values))
```

```python
# Initialize variables to track the best parameters and corresponding MAE
best_parameters = None
best_mae = float('inf')
```

```python
# Split the data into training and validation sets

split_point = int(len(weekly_data) * 0.8)  # 80% training, 20% validation
training_data = weekly_data[:split_point]
validation_data = weekly_data[split_point:]
```

```python
# Iterate through the parameter grid
for alpha, beta, gamma in parameter_grid:
    # Fit the Exponential Smoothing model on the training data
    model = sm.tsa.ExponentialSmoothing(training_data, trend='add', seasonal='add', seasonal_periods=52)
    fit = model.fit(smoothing_level=alpha, smoothing_trend=beta, smoothing_seasonal=gamma)
    
    # Generate forecasts for the validation period
    forecast_periods = len(validation_data)  # Forecast the entire validation period
    forecast_values = fit.forecast(steps=forecast_periods)
    
    # Calculate MAE for the forecasts using the validation data
    mae = np.mean(np.abs(forecast_values - validation_data))
    
    # Check if this set of parameters results in a lower MAE
    if mae < best_mae:
        best_mae = mae
    best_parameters = (alpha, beta, gamma)
```

```python
# Print the best parameters and corresponding MAE
print(f"Best Parameters (alpha, beta, gamma): {best_parameters}")
print(f"Best MAE: {best_mae}")
```

#### Alpha (Level Smoothing): 0.5
#### Beta (Trend Smoothing): 0.2
#### Gamma (Seasonality Smoothing): 0.2

#### MAE: .4463, which indicates the accuracy of our Exponential Smoothing model's forecasts on the validation dataset.
